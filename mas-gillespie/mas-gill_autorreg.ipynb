```python
rhlife = 120 #s = 2min
phlife = 3600 #s = 1h
gammar = np.log(2)/rhlife #/s
gammap = np.log(2)/phlife #/s
b = 10
kp = b*gammar

p_unrep = 1200
kr_max = p_unrep*gammar*gammap/kp #/s
r0 = 0
p0 = 0

kd = 800
n = 2
```

```python
def Step(t_old, r_old, p_old):
    
    kr = kr_max/(1.0+(p_old/kd)**n)

    K = kr + gammar*r_old + kp*r_old + gammap*p_old
    
    Dt = np.log(1.0/random.random())/K
    which = np.random.random()

    t_new = t_old + Dt
    
    if(which < kr/K):
        r_new = r_old + 1
        p_new = p_old
    elif(which >= kr/K and which < (kr+gammar*r_old)/K):
        r_new = r_old - 1
        p_new = p_old
    elif(which >= (kr+gammar*r_old)/K and which < (kr+gammar*r_old+kp*r_old)/K):
        r_new = r_old
        p_new = p_old + 1
    else:
        r_new = r_old
        p_new = p_old - 1
        
    return t_new, r_new, p_new
```

```python
t_total = 10*phlife
dt = 0.01

r, p = cell(dt, t_total)
```
